```python
# import the necessary packakges
import numpy as np
import matplotlib.pyplot as plt

# defining the functions for equations and their derivatives
def function_1d(x):
    return x**2 - 2*x + 2

def derivative_1d(x):
    return 2*x - 2

def function_2d(x, y):
    return (1 - x)**2 + 100*(y - x**2)**2

def derivative_2d(x, y):
    dx = -1*(1 - x) - 400*x*(y - x**2)
    dy = 200*(y - x**2)
    return dx, dy
```

```python
# gradient descent for the first equation
# actual equation: f(x) = x^2 - 2x + 2
# alpha = learning rate

def gradient_1d(alpha, precision):
    x_min = np.random.uniform(-5, 5)
    iterations = 0
    x_vals = []
    y_vals = []

    while True:
        gradient = derivative_1d(x_min)
        next_x = x_min - alpha * gradient

        x_vals.append(next_x)
        y_vals.append(function_1d(next_x))

        if abs(next_x - x_min) < precision:
            break

        x_min = next_x
        iterations += 1

    return x_vals, y_vals, iterations
```

```python
# gradient descent for the second equation
# actual equation: f(x, y) = (1 - x)^2 + 100(y - x^2)^2

def gradient_2d(alpha, precision):
    x_min = np.random.uniform(-2, 2)
    y_min = np.random.uniform(-2, 2)
    iterations = 0
    x_vals = []
    y_vals = []
    z_vals = []

    while True:
        gradient_x, gradient_y = derivative_2d(x_min, y_min)
        next_x = x_min - alpha * gradient_x
        next_y = y_min - alpha * gradient_y

        x_vals.append(next_x)
        y_vals.append(next_y)
        z_vals.append(function_2d(next_x, next_y))

        if abs(next_x - x_min) < precision and abs(next_y - y_min) < precision:
            break

        x_min = next_x
        y_min = next_y
        iterations += 1

    return x_vals, y_vals, z_vals, iterations
```

```python
# set hyperparameters
alpha = 0.001
precision = 0.0001
```

```python
# gradient descent for the first equation
x_vals_1d, y_vals_1d, iterations_1d = gradient_1d(alpha, precision)

# gradient descent for the second equation
x_vals_2d, y_vals_2d, z_vals_2d, iterations_2d = gradient_2d(alpha, precision)
```

```python
#print the "NUMBERS"
print("First Equation:")
print(f"Number of iterations: {iterations_1d}")

print("\nSecond Equation:")
print(f"Number of iterations: {iterations_2d}")
```

```python
# Visualizing the results for the first eqution
plt.figure(figsize=(10, 6))
plt.plot(x_vals_1d, y_vals_1d, color='blue', marker='*', linestyle='-')
plt.title("Gradient Descent for the fisrt equation")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.grid(True)
plt.show()
```

```python
#Visualizing the results for the second equation
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x_vals_2d, y_vals_2d, z_vals_2d, marker='o', linestyle='-')
ax.set_title("Gradient Descent for the Second Equation")
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.show()
```
